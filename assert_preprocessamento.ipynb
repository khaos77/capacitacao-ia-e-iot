```python
from seaborn import load_dataset

data = load_dataset("titanic")
data.info()
```

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer

data = data.dropna()
data.isnull().sum()
data
```

```python
data = data.drop('class', axis = 1)
data
```

```python
y = data['survived']
x = data.drop('survived', axis = 1)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


column_transformer = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), ['sex', 'embarked', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone'])],
    remainder='passthrough'
)

df_transformed = column_transformer.fit_transform(x)
df_transformed = pd.DataFrame(data=df_transformed, columns=list(column_transformer.get_feature_names_out()))
df_transformed
```

```python
df_transformed.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

X = df_transformed

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)




model_1 = DecisionTreeClassifier(criterion = "entropy", max_depth=3)
model_2 = DecisionTreeClassifier(criterion = "gini")


model_1.fit(X_train, y_train)
model_2.fit(X_train, y_train)




result_1 = model_1.predict(X_test)
acc1 = accuracy_score(result_1, y_test)


result_2 = model_2.predict(X_test)
acc2 = accuracy_score(result_2, y_test)
print("model 1: ", acc1*100, "%")
print("model 2: ", acc2*100, "%")
```

```python
from sklearn.neighbors import KNeighborsClassifier

neigh_1 = KNeighborsClassifier(n_neighbors=7)
neigh_1.fit(X_train, y_train)

neigh_2 = KNeighborsClassifier(n_neighbors=5)
neigh_2.fit(X_train, y_train)

result_1 = neigh_1.predict(X_test)
acc1 = accuracy_score(result_1, y_test)


result_2 = neigh_2.predict(X_test)
acc2 = accuracy_score(result_2, y_test)
print("neigh 1: ", acc1*100, "%")
print("neigh 2: ", acc2*100, "%")

```
